# Neighbour joining tree of full Y chromosome dataset

```r
library(phangorn)
library(lemon)

library(ggplot2)
library(ggtree)
library(treeio)

source("utils/utils.R")

colours <- c('#1E74CB', '#EFAD15', '#8B0017')

pops <- c("Neandertal", "African", 'non-African')

aln_full <- read.phyDat("data/output_data/data_full_merged.hg19_mapped.fa", format = "fasta")
aln_full <- as.DNAbin(aln_full)

buildtree_distdna_midpoint <- function(x) midpoint(nj(dist.dna(x, model = "raw", pairwise.deletion = T)))

tr <- buildtree_distdna_midpoint(aln_full)

bp <- boot.phylo(tr, aln_full, buildtree_distdna_midpoint, B = 1)

tr <- fix_tree_labels(tr)
tr <- as.treedata(tr, bp)
tr <- drop.tip(tr, tip = "chimp")

options(repr.plot.width = 24, repr.plot.height = 22, repr.plot.res = 300)

p <- ggtree(tr, ladderize = F) +
    #  geom_text2(aes(label=bootstrap), hjust=-0.5, vjust=0.5, size=3) +
     coord_capped_cart(gap = 1, x = c(0.0001, 0.016))

p$data$label[p$data$label == 'A00 (Shum Laka)'] = 'A00 (Shum Laka, ~8 ka)'

tree_metadata_temp <- data.frame(lapply(tree_metadata, function(x) {
                  gsub('A00 (Shum Laka)', 'A00 (Shum Laka, ~8 ka)', x, fixed=TRUE)
              }))

p <- p %<+% tree_metadata_temp +
      geom_tiplab(aes(color = group), size = 12, show.legend = FALSE, align=TRUE, linetype='58', hjust = 0) +
      scale_color_manual(values=colours, breaks=pops) +
      geom_polygon(aes(fill = group, x = 0, y = 0)) +
      scale_fill_manual(breaks=pops, values=colours, na.translate = F) + 
      theme(legend.position = c(0.12,0.65), legend.title = element_blank(),  legend.text = element_text(size=40), legend.key.size = unit(1.5, 'cm')) +
      scale_y_reverse()

p
```

# Full BEAST tree based on divergence filtered dataset

```r
library(phangorn)
library(lemon)

library(ggplot2)
library(ggtree)
library(treeio)

source("utils/utils.R")

colours <- c('#1E74CB', '#EFAD15', '#8B0017')
pops <- c("Neandertal", "African", 'non-African')

tr <- read.beast('BEAST/full_runs/data_full_divergence_filtered/data_full_divergence_filtered.fixed_skyline.combined.treeannotator')
tr <- drop.tip(tr, tip = "chimp")

tr@phylo <- fix_tree_labels(tr@phylo)

options(repr.plot.width = 28, repr.plot.height = 18, repr.plot.res = 300)

p <- ggtree(tr, ladderize = T) +
      scale_y_reverse() +
      xlim(-10000, 475000) +
    #   geom_nodelab(aes(label=node), vjust=-.5, size=6, hjust = 1.2)
      geom_nodelab(aes(label=paste(round(height/1000, 0), ' ka\n[', round(range_lower(height_0.95_HPD)/1000, 0), ', ', round(range_upper(height_0.95_HPD)/1000, 0), ']' , sep = ""),
      subset=(node %in% c(35, 36, 37, 67, 38, 40, 41, 42, 44))), vjust=+0.5, size=9, hjust = 1, nudge_x=-5000)

temp = p$data$label == "Chagyrskaya2"
temp[is.na(temp)] <- FALSE
chag2_str = paste('Chagyrskaya2, ', round(p$data[temp,]$height_median/1000, 0), ' ka [', round(p$data[temp,]$height_0.95_HPD[[1]][1]/1000, 0), ', ', round(p$data[temp,]$height_0.95_HPD[[1]][2]/1000, 0), ']' , sep = "")
p$data$label[temp] = chag2_str

p$data$label[p$data$label == 'A00 (Shum Laka)'] = 'A00 (Shum Laka, ~8 ka)'

tree_metadata_temp <- data.frame(lapply(tree_metadata, function(x) {
                  gsub("Chagyrskaya2", chag2_str, x, fixed=TRUE)
              }))

tree_metadata_temp <- data.frame(lapply(tree_metadata_temp, function(x) {
                  gsub('A00 (Shum Laka)', 'A00 (Shum Laka, ~8 ka)', x, fixed=TRUE)
              }))

p <- p %<+% tree_metadata_temp +     
      geom_tiplab(aes(color = group), size = 11, show.legend = FALSE, align=TRUE, linetype='58') +
      scale_color_manual(values=colours, breaks=pops) +
      geom_polygon(aes(fill = group, x = 0, y = 0)) +
      scale_fill_manual(breaks=pops, values=colours, na.translate = F) + 
      theme(legend.position=c(0.125,0.65), legend.title=element_blank(),
      legend.text = element_text(size=40), legend.key.size = unit(1.5, 'cm'))

p
```

# BEAST tree without present-day A00 - filtered

```r
library(phangorn)
library(lemon)

library(ggplot2)
library(ggtree)
library(treeio)

source("utils/utils.R")

colours <- c('#1E74CB', '#EFAD15', '#8B0017')
pops <- c("Neandertal", "African", 'non-African')

tr <- read.beast('BEAST/full_runs/shum_laka_age_estimation/filtered_data/data_full_no_A00_divergence_filtered.fixed_skyline.combined.treeannotator')
tr <- drop.tip(tr, tip = "chimp")
tr <- drop.tip(tr, tip = "Chagyrskaya2")
tr <- drop.tip(tr, tip = "Mezmaiskaya2")

tr@phylo <- fix_tree_labels(tr@phylo)

options(repr.plot.width = 28, repr.plot.height = 12, repr.plot.res = 300)

p <- ggtree(tr, ladderize = T) +
      scale_y_reverse() +
      xlim(-10000, 340000) +
    #   geom_nodelab(aes(label=node), vjust=-.5, size=6, hjust = 1.2)
      geom_nodelab(aes(label=paste(round(height/1000, 0), ' ka\n[', round(range_lower(height_0.95_HPD)/1000, 0), ', ', round(range_upper(height_0.95_HPD)/1000, 0), ']' , sep = ""),
      subset=(node %in% c(32, 33, 34, 36, 37, 39))), vjust=+0.5, size=9, hjust = 1, nudge_x=-5000)

temp = p$data$label == "A00 (Shum Laka)"
temp[is.na(temp)] <- FALSE
a00_str = paste('A00 (Shum Laka), ', round(p$data[temp,]$height_median/1000, 0), ' ka [', round(p$data[temp,]$height_0.95_HPD[[1]][1]/1000, 0), ', ', round(p$data[temp,]$height_0.95_HPD[[1]][2]/1000, 0), ']' , sep = "")
p$data$label[temp] = a00_str

tree_metadata_temp <- data.frame(lapply(tree_metadata, function(x) {
                  gsub("A00 (Shum Laka)", a00_str, x, fixed=TRUE)
              }))

p <- p %<+% tree_metadata_temp +
      scale_color_manual(values=colours, breaks=pops) +
      geom_tiplab(aes(color = group, subset=node>1), size = 11, show.legend = FALSE, align=TRUE, linetype='58') +
      geom_tiplab(aes(color = group, subset=node==1, label=a00_str), size = 11, show.legend = FALSE, align=TRUE, linetype='58', fontface='bold')

p <- p %>% scaleClade(37, 0.3) %>%
    collapse(node = 37, "max")

p
```
